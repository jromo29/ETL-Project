```python
import pandas as pd
import requests
import time
from pprint import pprint
from pandas import ExcelWriter

```

```python
file_to_load = 'SPLG_All_Holdings.xls'
LgCap_data = pd.read_excel(file_to_load, header=3, sheet="SPY_All_Holdings")
LgCap_data.head(20)
```

```python
#Identify DataTypes
LgCap_data.dtypes
```

```python
# Add Column Name with fund name
LgCap_data["FundName"] = "SPY_LargeCap"
LgCap_data.head(20)
```

```python
file_to_load2 = 'SLY_All_Holdings.xls'
SmCap_data = pd.read_excel(file_to_load2, header=3, sheet="SLY_All_Holdings")
SmCap_data.head(20)
```

```python
SmCap_data.dtypes
```

```python
SmCap_data["FundName"] = "SPY_SmallCap"
SmCap_data.head(20)
```

```python
file_to_load3 = 'MDY_All_Holdings.xls'
MdCap_data = pd.read_excel(file_to_load3, header=3, sheet="MDY_All_Holdings")
MdCap_data.head(20)
```

```python
MdCap_data.dtypes
```

```python
MdCap_data["FundName"] = "SPY_MediumCap"
MdCap_data.head(20)
```

```python
#Combine all dataframes
AllStocks_df= pd.concat([SmCap_data, LgCap_data])

#Reset Index
AllStocks_df = AllStocks_df.reset_index()
AllStocks_df

#Ensure there are no null values
print(len(AllStocks_df.Identifier))
AllStocks_df.isnull().sum()

```

```python
#Obtain Current stock price for all listed stocks



close = []
high = []
low = []
openprice = []

total_stocks = len(AllStocks_df.Identifier)

for x in range(0,total_stocks):

    stock = AllStocks_df.Identifier[x]
    
    url = "https://api.iextrading.com/1.0/stock/" + stock +"/ohlc"
    
    response = requests.get(url).json()
    
    close.append(response['close']['price'])
    high.append(response['high'])
    low.append(response['low'])
    openprice.append(response['open']['price'])
    
    time.sleep(.5)
    
```

```python
#Add columns to dataframe

AllStocks_df['open'] = openprice
AllStocks_df['high'] = high
AllStocks_df['low'] = low
AllStocks_df['close'] = close


AllStocks_df.drop(columns="index", axis=1, inplace=True)
AllStocks_df.head()
```

```python
#Obtain key stats for all listed stocks
marketcap = []
beta = []
wk52high = []
wk52low = []
EPS = []

total_stocks = len(AllStocks_df.Identifier)

for x in range(0,total_stocks):

    stock = AllStocks_df.Identifier[x]
    
    url = "https://api.iextrading.com/1.0/stock/" + stock +"/stats"
    
    response = requests.get(url).json()
    
    marketcap.append(response['marketcap'])
    beta.append(response['beta'])
    wk52high.append(response['week52high'])
    wk52low.append(response['week52low'])
    EPS.append(response['ttmEPS'])
    
    time.sleep(.5)
    
print(marketcap)
    
```

```python
AllStocks_df['marketcap'] = marketcap
AllStocks_df['beta'] = beta
AllStocks_df['wk52high'] = wk52high
AllStocks_df['wk52low'] = wk52low
AllStocks_df['EPS'] = EPS

AllStocks_df.head()
```

```python
AllStocks_df.to_excel("StockData_Output.xlsx")
```
